```python
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
plt.rcParams['figure.figsize'] = 8,4
import warnings
warnings.filterwarnings('ignore')
```

```python
#Distribution:
vis1 = sns.distplot(stats['Internet_Users'])
```

```python
vis1 = sns.distplot(stats['Internet_Users'], bins = 30)
```

```python
#Boxplots
vis2 = sns.boxplot(data = stats, x = 'Income_Group', y = 'Birth_Rate')
```

```python
#Seaborn Gallery https://seaborn.pydata.org/examples/index.html
```

```python
vis3 = sns.lmplot(data = stats, x = 'Internet_Users', y = 'Birth_Rate', \
                  fit_reg = False, hue = 'Income_Group', size = 5) #linear relationship plot
```

```python
#Keyword Arguments
```

```python
#marker size
vis3 = sns.lmplot(data = stats, x = 'Internet_Users', y = 'Birth_Rate', \
                  fit_reg = False, hue = 'Income_Group', size = 8, \
                  scatter_kws = {'s':100}) #linear relationship plot, scatter_kws uses a dictionary from pyplot
```
